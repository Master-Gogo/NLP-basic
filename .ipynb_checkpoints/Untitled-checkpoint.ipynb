```python
# using simple python function
import numpy as np

def bag_of_words(texts):
    # create an empty vocabulary set
    vocabulary = set()

    # loop over each text in the corpus
    for text in texts:
        # split the text into words
        words = text.split()
        # add the words to the vocabulary set
        vocabulary.update(words)

    # create an empty bag-of-words matrix
    bow = np.zeros((len(texts), len(vocabulary)))

    # loop over each text in the corpus
    for i, text in enumerate(texts):
        # split the text into words
        words = text.split()
        # loop over each word in the vocabulary
        for j, word in enumerate(vocabulary):
            # count the number of occurrences of the word in the text
            count = words.count(word)
            # update the bag-of-words matrix
            bow[i, j] = count

    return bow

bow_representation = bag_of_words(shuffled_df_bow['review'])

# print the bag-of-words representation for each sentence in the corpus
for i in range(len(shuffled_df_bow)):
    print("Sentence ", i+1, ": ", bow_representation[i])
    break
```

# IF-IDF

```python
import math
from collections import Counter

def calculate_tf(word_dict, word_count):
    tf_dict = {}
    for word, count in word_dict.items():
        tf_dict[word] = count / float(word_count)
    return tf_dict

def calculate_idf(documents, word):
    N = len(documents)
    n = sum(1 for document in documents if word in document)
    if n > 0:
        return math.log(N / float(n))
    else:
        return 0.0

def calculate_tfidf(documents):
    tfidf_documents = []
    for document in documents:
        word_count = len(document)
        word_dict = Counter(document)
        tf_dict = calculate_tf(word_dict, word_count)
        tfidf_dict = {}
        for word, tf in tf_dict.items():
            idf = calculate_idf(documents, word)
            tfidf_dict[word] = tf * idf
        tfidf_documents.append(tfidf_dict)
    return tfidf_documents

# Sample review
review = "I loved this restaurant! The food was delicious and the service was excellent. but i loved this"

# Tokenize the review
tokens = review.split()

# Calculate TF-IDF for the review
tfidf_review = calculate_tfidf([tokens])

# Print the extracted features
print(tfidf_review)
```

```python
import torch
import numpy as np
from transformers import BertTokenizer, BertForSequenceClassification, AdamW

# Define some hyperparameters
MAX_LEN = 128
BATCH_SIZE = 32
LEARNING_RATE = 2e-5
NUM_EPOCHS = 2
MODEL_PATH = "bert_review_classifier.pt"

# Load the pre-trained BERT tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased', do_lower_case=True)

# Load the sample reviews data
reviews = [
    ("This movie was great!", 1),
    ("The film was okay.", 0),
    ("I didn't enjoy the movie at all.", 0),
    ("The acting was superb in this film.", 1),
    ("The plot was confusing and hard to follow.", 0)
]
texts = [text for text, label in reviews]
labels = [label for text, label in reviews]

# Tokenize the texts and convert them to input features for BERT
input_ids = []
attention_masks = []
for text in texts:
    # Tokenize the text
    tokens = tokenizer.tokenize(text)
    if len(tokens) > MAX_LEN - 2:
        tokens = tokens[:MAX_LEN - 2]

    # Add the special tokens
    tokens = ["[CLS]"] + tokens + ["[SEP]"]
    input_id = tokenizer.convert_tokens_to_ids(tokens)

    # Pad or truncate the input sequence to the maximum length
    padding_length = MAX_LEN - len(input_id)
    input_id += [0] * padding_length
    attention_mask = [1] * len(input_id)

    # Convert the input features to tensors and append them to the lists
    input_ids.append(input_id)
    attention_masks.append(attention_mask)

# Convert the input features and labels to tensors
input_ids = torch.tensor(input_ids)
attention_masks = torch.tensor(attention_masks)
labels = torch.tensor(labels)

# Create a TensorDataset containing the input features and labels
dataset = torch.utils.data.TensorDataset(input_ids, attention_masks, labels)

# Split the dataset into a training set and a validation set
train_size = int(0.8 * len(dataset))
val_size = len(dataset) - train_size
train_dataset, val_dataset = torch.utils.data.random_split(dataset, [train_size, val_size])

# Create data loaders for the training set and validation set
train_dataloader = torch.utils.data.DataLoader(train_dataset, sampler=torch.utils.data.RandomSampler(train_dataset), batch_size=BATCH_SIZE)
val_dataloader = torch.utils.data.DataLoader(val_dataset, sampler=torch.utils.data.SequentialSampler(val_dataset), batch_size=BATCH_SIZE)

# Load the pre-trained BERT model for sequence classification
model = BertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=2)

# Define the optimizer and the loss function
optimizer = AdamW(model.parameters(), lr=LEARNING_RATE, eps=1e-8)
loss_fn = torch.nn.CrossEntropyLoss()

# Train the model
for epoch in range(NUM_EPOCHS):
    # Set the model to training mode
    model.train()

    # Initialize variables for tracking training loss and number of batches
    train_loss = 0.0
    num_batches = 0

    # Loop over the batches in the training set
    for batch in train_dataloader:
        # Extract the input features and labels from the batch
        batch_input_ids = batch[0]
        batch_attention_masks = batch[1]
        batch_labels = batch[2]

        # Zero the gradients
        optimizer.zero_grad()

        # Forward pass
        outputs = model(input_ids=batch_input_ids, attention_mask=batch_attention_masks, labels)
                # Calculate the loss
        loss = outputs[0]

        # Backward pass
        loss.backward()

        # Update the model parameters
        optimizer.step()

        # Update the training loss and number of batches
        train_loss += loss.item()
        num_batches += 1

    # Calculate the average training loss for the epoch
    avg_train_loss = train_loss / num_batches

    # Print the average training loss for the epoch
    print(f"Epoch {epoch + 1} / {NUM_EPOCHS} - Average training loss: {avg_train_loss:.2f}")

    # Set the model to evaluation mode
    model.eval()

    # Initialize variables for tracking validation loss and number of batches
    val_loss = 0.0
    num_batches = 0

    # Loop over the batches in the validation set
    for batch in val_dataloader:
        # Extract the input features and labels from the batch
        batch_input_ids = batch[0]
        batch_attention_masks = batch[1]
        batch_labels = batch[2]

        # Forward pass
        with torch.no_grad():
            outputs = model(input_ids=batch_input_ids, attention_mask=batch_attention_masks, labels=batch_labels)

        # Calculate the loss
        loss = outputs[0]

        # Update the validation loss and number of batches
        val_loss += loss.item()
        num_batches += 1

    # Calculate the average validation loss for the epoch
    avg_val_loss = val_loss / num_batches

    # Print the average validation loss for the epoch
    print(f"Epoch {epoch + 1} / {NUM_EPOCHS} - Average validation loss: {avg_val_loss:.2f}")

# Save the model
torch.save(model.state_dict(), MODEL_PATH)

# Load the saved model
model = BertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=2)
model.load_state_dict(torch.load(MODEL_PATH))

# Set the model to evaluation mode
model.eval()

# Tokenize the sample reviews and convert them to input features for BERT
sample_reviews = ["This movie was great!", "The film was okay.", "I didn't enjoy the movie at all."]
sample_input_ids = []
sample_attention_masks = []
for text in sample_reviews:
    # Tokenize the text
    tokens = tokenizer.tokenize(text)
    if len(tokens) > MAX_LEN - 2:
        tokens = tokens[:MAX_LEN - 2]

    # Add the special tokens
    tokens = ["[CLS]"] + tokens + ["[SEP]"]
    input_id = tokenizer.convert_tokens_to_ids(tokens)

    # Pad or truncate the input sequence to the maximum length
    padding_length = MAX_LEN - len(input_id)
    input_id += [0] * padding_length
    attention_mask = [1] * len(input_id)

    # Convert the input features to tensors and append them to the lists
    sample_input_ids.append(input_id)
    sample_attention_masks.append(attention_mask)

# Convert the input features to tensors
sample_input_ids = torch.tensor(sample_input_ids)
sample_attention_masks = torch.tensor(sample_attention_masks)

# Create a TensorDataset containing the input features and labels
sample_dataset = torch.utils.data.TensorDataset(sample_input_ids, sample_attention_masks)

# Create a data loader for the sample dataset
sample_dataloader = torch.utils.data.DataLoader(sample_dataset, sampler=torch.utils.data.SequentialSampler(sample_dataset), batch_size=BATCH_SIZE)

# Make predictions on the sample dataset
for batch in sample_dataloader:
    # Extract the input features from the batch
    batch_input_ids = batch[0]
    batch_attention_masks = batch[1]

    # Forward

```
